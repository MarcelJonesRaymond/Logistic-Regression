```python
def right_prediction(actual,predicted):
    a=pd.DataFrame()
    for i in np.sort(actual.unique()):
        try:
            a.loc[i,"Total instance"]=len(actual[actual==i])
            a.loc[i,"right_identification"]=pd.Series(predicted[actual == i]).value_counts()[i]
            a.loc[i, '%'] = round(100*pd.Series(predicted[actual == i]).value_counts()[i]/len(actual[actual == i]), 1)
        except:
            a.loc[i, '%'] =0
    return a
```

```python
def con_matrix(actual, predictions):
    from sklearn.metrics import confusion_matrix

    matrix = pd.DataFrame(confusion_matrix(actual, predictions), columns = np.unique(actual), index = np.unique(actual))
    # confusion_matrix() returns the confusion matrix. Wrapping this matrix in a dataframe with appropriate column names and index
    matrix.loc[:,'total_count_actual_class'] = matrix.sum(axis = 1)  # sum accross the columns
    matrix.loc['total_count_predicted'] = matrix.sum(axis = 0)  # row sum
    return matrix
```

```python
con_matrix(y_test, yhat)
```

# knn

```python
from sklearn.neighbors import KNeighborsClassifier
score=[]
n=list(range(1,51))
neighbors=list(filter(lambda x: x%2!=0,n))
for i  in neighbors:
    
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    score.append(knn.score(x_train,y_train))
```

```python
plt.plot(neighbors,score,marker="^")
plt.xticks(range(1,51))
plt.xticks(rotation=90)
```

```python
mse=[1-i for i in score]
```
